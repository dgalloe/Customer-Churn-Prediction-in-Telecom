```python
# Renombrar columnas a minusculas y gui√≥n bajo entre palabras

columns_contracts = data_contract.columns

# Creamos variable en blanco para almacenar el resultado de la siguiente funci√≥n
columns_contracts_low = []

#Funci√≥n para agregar gui√≥n bajo antes de encontrar mayusculas en medio del nombre de la columna
def add_scripts(columns_contracts_low):
    result = ""
    # Agregamos esta bandera para evitar que se coloqu√© _ cada vez que encuentra mayusculas, solo queremos encontrar a primera mayuscula, colocar _ y 
    # que se detenga la funci√≥n. 
    uppercase = False
    for i, caracter in enumerate(columns_contracts_low):

        if i != 0 and caracter.isupper():
            # Solo agregamos gui√≥n si NO estamos ya en una secuencia de may√∫sculas
            if not uppercase:
                result += "_"
            uppercase = True  # Ahora estamos en una secuencia de may√∫sculas
        elif caracter.islower():
            uppercase = False  # Ya no estamos en may√∫sculas
        
        result += caracter
    
    return result.lower()

# Bucle FOR para aplicar funci√≥n add_scripts a cada elemento de columns_contracts
for column_name in columns_contracts:
    columns_contracts_low.append(add_scripts(column_name))

data_contract.columns = columns_contracts_low
data_contract.info()
```

```python
# Revisar la existencia de duplicados en todo el dataframe y si hay fechas y horas repetidas en la columna 'date'
print(f"{bold_start}N√∫mero de valores duplicados en el Dataframe Contracts:{bold_end}",data_contract.duplicated().sum())
print(f"{bold_start}N√∫mero de customer_id √∫nicos:{bold_end}",data_contract['customer_id'].nunique())
print(f"{bold_start}N√∫mero de begin_date √∫nicos:{bold_end}",data_contract['begin_date'].nunique())
print(f"{bold_start}End_date √∫nicos:{bold_end}",data_contract['end_date'].unique())
print(f"{bold_start}Type √∫nicos:{bold_end}",data_contract['type'].unique())
print(f"{bold_start}Paperless_billing √∫nicos:{bold_end}",data_contract['paperless_billing'].unique())
print(f"{bold_start}Payment_method √∫nicos:{bold_end}",data_contract['payment_method'].unique())
print(f"{bold_start}N√∫mero de monthly_charges  √∫nicos:{bold_end}",data_contract['monthly_charges'].nunique())
print(f"{bold_start}N√∫mero de total_charges √∫nicos:{bold_end}",data_contract['total_charges'].nunique())
```
